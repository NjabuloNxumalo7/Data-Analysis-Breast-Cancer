```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


#Data Loading

data = pd.read_csv('gbsg.csv') 

#Display basic information
print("Dataset Info:")
print(data.info())

print("\nNumber of Rows and Columns:", data.shape)
print("\nSummary Statistics:")
print(data.describe())

#Data Cleaning

print("\nMissing values in each column:")
print(data.isnull().sum())

#Handle missing values

data.fillna(data.mean(), inplace=True) 

#Check for duplicates
duplicates = data.duplicated().sum()
print("\nNumber of duplicate rows:", duplicates)

# Remove duplicates if any
data.drop_duplicates(inplace=True)


# Data Exploration

print("\nDescriptive statistics:")
print(data.describe())

# Visualize distributions
plt.figure(figsize=(8, 5))
sns.histplot(data['age'], kde=True, bins=10, color='skyblue')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(data['size'], kde=True, bins=10, color='lightcoral')
plt.title('Tumor Size Distribution')
plt.xlabel('Tumor Size')
plt.ylabel('Frequency')
plt.show()

# Data Manipulation

average_size_by_meno = data.groupby('meno')['size'].mean()
print("\nAverage Tumor Size by Menopause Status:")
print(average_size_by_meno)



print("\nSummary of Findings:")
print("This dataset consists of breast cancer patients with various details.")
print("We explored key attributes such as age, tumor size, and their relationships.")
print("Some interesting insights include the correlation between variables like age and tumor size.")

```
